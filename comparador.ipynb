```python
from tkinter import Label, Button, filedialog, messagebox, Entry, END, Tk, mainloop, Frame, LEFT, RIGHT, PhotoImage
import pandas as pd
```

```python
# Back-end
# ----------------------Upload-----------------------
def upload_arquivo(descricao_caminho1:Label, descricao_caminho2:Label):
  caminho_do_arquivo1 = filedialog.askopenfilename(filetypes=[("Excel files", "*.xls *.xlsx")])
  caminho_do_arquivo2 = filedialog.askopenfilename(filetypes=[("Excel files", "*.xls *.xlsx")])

  if caminho_do_arquivo1 and caminho_do_arquivo2:
    nome_arquivo1 = pega_nome(caminho_do_arquivo1)
    nome_arquivo2 = pega_nome(caminho_do_arquivo2)
    descricao_caminho1.config(text=f'{nome_arquivo1}')
    descricao_caminho2.config(text=f'{nome_arquivo2}')

  comparar_tabelas(caminho_do_arquivo1, caminho_do_arquivo2)
# ---------------------------------------------------

# ----------------------Nome do Arquivo-----------------------
def pega_nome(caminho_do_arquivo:str):
  diretorio = caminho_do_arquivo.split('/')

  return diretorio[-1]
# -------------------------------------------------------------

# -----------------------------------Filtro Zero KM-----------------------------------
def filtro_zerokm(descricao_caminho:Label):
  caminho = filedialog.askopenfilename(filetypes=[("Excel files", "*.xls *.xlsx")])

  tabela = pd.read_excel(caminho)

  filtro = tabela[tabela['Ano'] == 'Zero KM']

  if caminho:
    nome_arquivo = pega_nome(caminho)
    descricao_caminho.config(text=f'{nome_arquivo}')
    
  display(filtro)
# --------------------------------------------------------------------------------------

# ---------------------------------Comparar os arquivos---------------------------------
def comparar_tabelas(caminho1:str, caminho2:str):
  tabela1 = pd.read_excel(caminho1)
  tabela2 = pd.read_excel(caminho2)

  tabela1, tabela2 = tabela1.align(tabela2, join='outer', axis=1)  # Alinhar as colunas
  tabela1, tabela2 = tabela1.align(tabela2, join='outer', axis=0)  # Alinhar os índices

  comparacao = tabela1.compare(tabela2)

  display(comparacao)
```

```python
# FRONT-END
# -------------Configurações da Tela-----------------
janela = Tk()
largura = 550
altura = 230

icon = PhotoImage(file='imagem/logo.png')

janela.title('Comparador')
janela.iconphoto(False, icon)
janela.geometry(f'{largura}x{altura}')
janela.resizable(False, False)
janela.configure(bg='purple')
# ---------------------------------------------------

# ----------------------Labels-----------------------
label_espaco_inicial = Label(janela,
                             image=icon,
                             bg='purple')
label_espaco_inicial.grid(row=0,
                          column=1,
                          columnspan=4,
                          pady=3)

label_comparar1 = Label(janela,
                      text='Selecione um arquivo',
                      fg='white',
                      width=20,
                      height=3,
                      bg='purple')
label_comparar1.grid(row=1,
                  column=1,
                  pady=5,
                  padx=3)

label_comparar2 = Label(janela,
                      text='Selecione um arquivo',
                      fg='white',
                      width=20,
                      height=3,
                      bg='purple')
label_comparar2.grid(row=1,
                  column=2,
                  pady=5,
                  padx=3)

label_filtar = Label(janela,
                      text='Selecione um arquivo',
                      fg='white',
                      width=20,
                      height=3,
                      bg='purple')
label_filtar.grid(row=2,
                  column=1,
                  columnspan=2,
                  pady=5,
                  padx=3)
# ---------------------------------------------------

# ----------------------Botões-----------------------
btn_comparar = Button(janela,
                      text='Comparar',
                      width=15,
                      height=2,
                      bg='gray',
                      command=lambda: upload_arquivo(label_comparar1, label_comparar2))
btn_comparar.grid(row=1,
                 column=0,
                 pady=5)

btn_filtrar = Button(janela,
                      text='Filtrar',
                      width=15,
                      height=2,
                      bg='gray',
                      command=lambda: filtro_zerokm(label_filtar))
btn_filtrar.grid(row=2,
                 column=0)
# ---------------------------------------------------


janela.mainloop()
```
